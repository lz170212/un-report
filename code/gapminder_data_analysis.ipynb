```python
# Exercise: Find the most recent year in the DataFrame.

gapminder['year'].max()
```

```python
(
    gapminder
    .agg({'lifeExp':['mean', 'std'],
         'year':'max'})
)
```

```python
(
    gapminder
    .query('year == 2007')
    .agg({'lifeExp':'mean'})
)
```

```python
# Exercise: Find the mean gdpPercap for the first year in the dataset
year_min = gapminder['year'].min()
```

```python
# @
(
    gapminder
    .query('year == @year_min')
    .agg({'gdpPercap':'mean'})
    .round(decimals = 2)
)
```

```python
# another way
x = (
    gapminder
    .query('year == year.min()')
    .agg({'gdpPercap':['mean', 'std']})
    .round(decimals = 2)
)
```

```python
# multiple conditions
# %%timeit
(
    gapminder
    .query('country == "United States" and year > 2000')
)
```

```python
country_list = ["United States", "Canada"]
(
    gapminder 
    .query('country in @country_list and year > 2000')
)
```

```python
country_list = ["United States", "Canada"]
(
    gapminder 
    .query('country not in @country_list')
    .query('year == 2007')
    .query('continent == "Americas"')
)
```

```python
# Sort data with sort_values()
```

```python
(
    gapminder
    .query('year == 2007')
    .groupby('continent')
    .agg({'lifeExp':'mean'})
    .sort_values('lifeExp', ascending = False)
)
```

```python
# Make new variables with assign()
```

```python
gapminder['gdp'] = gapminder['gdpPercap'] * gapminder['pop']
```

```python
# We can also do this by ...
(
    gapminder
    .assign(gdp = lambda df: df['gdpPercap'] * df['pop'],
            popInMillions = lambda df: df['pop']/1000000) # 1_000_000
)
```

```python
# Changing the shape of data
```

```python
# melt and pivot
(
    gapminder
    .filter(['continent', 'country', 'lifeExp', 'year'])
    .pivot(columns = 'year', values = 'lifeExp', index = ['continent', 'country']) 
)
```

```python
# melt and pivot
(
    gapminder
    .filter(['continent', 'country', 'lifeExp', 'year'])
    .pivot(columns = 'year', values = 'lifeExp', index = ['continent', 'country'])
    .reset_index()
    .melt(id_vars = ['continent', 'country'],
          value_name = 'lifeExp')
)
```
